```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.datasets import load_breast_cancer
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

Zad 1

```python
print(load_breast_cancer().data)
print(load_breast_cancer().target)
```

```python
dataset = load_breast_cancer()
data = load_breast_cancer().data
target = load_breast_cancer().target
data
```

```python
data.shape
```

```python
(values, count)=np.unique(target, return_counts = True)
print(values)
print(count)
```

```python
df_cancer=pd.DataFrame(data, columns=dataset.feature_names)
df_cancer['target'] = pd.Series(target)
```

Decyzja, które dane dostarczają najmniej inforamcji jest trudna, ponieważ może zaistnieć sytuacja, w której te dane okażą się niezbędne. Zakładając jednak, że chcemy je ograniczyć, wybrałbym kolumny z wartościami średnimi oraz najgorszymi z zebranych materiałów.

```python
df_copy=df_cancer.copy(deep=True)
df_copy = df_copy[df_copy.columns.drop(list(df_copy.filter(regex = 'error')))]
pd.set_option('display.max_rows', 10)
pd.set_option('display.max_columns', None)
df_copy
```

```python
df_copy.to_csv(r'dateset_cut.csv', index=False, sep=';')
```

W wyniku uruchomienia powyższego kodu otrzymaliśmy zredukowany zbiór danych, ale do dalszych działań wykorzystaliśmy pełen, ponieważ wyniki mogą okazać się dokładniejsze.


Zad 2

```python
data_ready=StandardScaler().fit_transform(data)
pca=PCA(n_components = 5)
components=pca.fit_transform(data_ready)
df = pd.DataFrame(data = components, columns = ['COMP1', 'COMP2', 'COMP3', 'COMP4', 'COMP5'])
df['TARGET'] = df_cancer['target']
df.to_csv(r'dateset_pca_5.csv', index=False, sep=';')
df
```

```python
df.var()
```

```python
df.var().sum()
```

```python
pca.explained_variance_ratio_.sum()
```

Otrzymane wyniki zostaną wykorzystane do porównania z innymi rozwiązaniami

Zad 3

```python
pca = PCA(svd_solver='auto', n_components=0.90)
components = pca.fit_transform(data_ready)
df = pd.DataFrame(data = components, columns = ['COMP1', 'COMP2', 'COMP3', 'COMP4', 'COMP5', 'COMP6', 'COMP7'])
df['TARGET'] = df_cancer['target']
df.to_csv(r'dateset_pca_n.csv', index=False, sep=';')
df
```

```python
df.var()
```

```python
df.var().sum()
```

```python
pca.explained_variance_ratio_.sum()
```

Do dobrania odpowiednich ilości komponentów użyliśmy svd_solver z parametrem auto. Wyszło, że dla minimum 90% wariacji potrzeba 7 parametrów. Poskutkowało to wzrostem wariancji wyjaśnionej dla wygenerowanych składowych i wariancji zbioru danych po redukcji.

Zad 4

```python
from sklearn.decomposition import TruncatedSVD
```

```python
svd = TruncatedSVD(n_components=5, n_iter=5, random_state=42)
a=svd.fit_transform(data_ready)
df = pd.DataFrame(data = a, columns = ['COMP1', 'COMP2', 'COMP3', 'COMP4', 'COMP5'])
df['TARGET'] = df_cancer['target']
df.to_csv(r'dateset_algorithm.csv', index=False, sep=';')
df
```

```python
df.var()
```

```python
df.var().sum()
```

```python
svd.explained_variance_ratio_.sum()
```

Wybraliśmy algorytm TruncatedSVD (innaczej LSA) z powodu chęci przetestowania i porównania z PCA. Przy wybranych parametrach dla 5 składowych otrzymaliśmy wyniki bardzo zbiżone do PCA dla takiej samej ilości. Może to być spowodowane tym, że oba algorytmy wykorzystują SVD
